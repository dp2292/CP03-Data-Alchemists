```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('previous_data.csv')
```

```python
df.head()
```

```python
#from all row in opponent column remove the first 3 character
df['Opposition'] = df['Opposition'].str[2:]
```

```python
df.head()
```

```python
Teams = ['Afghanistan', 'Australia', 'Bangladesh', 'England', 'India', 'New Zealand', 'Pakistan', 'South Africa', 'Sri Lanka', 'Ireland']

#remove row if does not contain any of the teams

df = df[df['Team'].isin(Teams) & df['Opposition'].isin(Teams)]
```

```python
#keep only Team and Opposition and % column

df = df[['Team', 'Opposition','Mat','Won','Lost', '%']]
```

```python
#export this data to csv file

df.to_csv('previous_data_per.csv', index=False)
```
