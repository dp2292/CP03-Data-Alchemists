```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
bow = pd.read_csv('bowling_avg.csv')
```

```python
bow.head()
```

```python
bow.drop(['Span'], axis=1, inplace=True)
```

```python
#remove odd rows

bow.drop(bow.index[1::2], inplace=True)

bow.head()
```

```python
#replece '-' with 0

bow.replace('-', 0, inplace=True)

bow.head()
```

```python
#remove Ct, St, 5, 10

bow.drop(['Ct', 'St', '5', '10'], axis=1, inplace=True)

bow.head()
```

```python
bow.drop(['BBI'], axis=1, inplace=True)

bow.head()
```

```python
bow.drop(['SR'], axis=1, inplace=True)

bow.head()
```

```python
#rename column

bow.rename(columns={'Player': 'player'}, inplace=True)

bow.rename(columns={'Mat': 'matches'}, inplace=True)

bow.rename(columns={'Balls': 'balls'}, inplace=True)

bow.rename(columns={'Mdns': 'maiden'}, inplace=True)

bow.rename(columns={'Runs': 'runs'}, inplace=True)

bow.rename(columns={'Wkts': 'wickets'}, inplace=True)

bow.rename(columns={'Ave': 'avg'}, inplace=True)

bow.rename(columns={'Econ': 'economy'}, inplace=True)


bow.head()
```

```python
#export to csv

bow.to_csv('bowling_avg_cleaned.csv', index=False)
```
