```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_backup = pd.read_csv('./batting_avg.csv')
df = data_backup.copy()
df.columns
```

```python
df.head()
```

```python
#remove row with missing values
df = df.dropna()
df.head()
```

```python
df.shape
```

```python
#replace - with 0
df = df.replace('-', 0)
df.head()
```

```python
#change column names

df.columns = ['Player', 'Span', 'matches', 'innings', 'not_out', 'runs', 'highest_score', 'avg', 'strike_rate', '100s', '50s', '0s']
df.head()
```

```python
df.drop(['Span'], axis=1, inplace=True)
df.head()
```

```python
#export to csv

df.to_csv('batting_avg_cleaned.csv', index=False)
```
