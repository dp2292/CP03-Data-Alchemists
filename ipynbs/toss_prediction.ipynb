```python
import math
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('../matches.csv')
```

```python
#keep only season, team1, team2, toss_winner, toss_decision, winner
df1 = df[['team1', 'team2','venue', 'toss_winner', 'toss_decision']]
df1.info()
```

```python
#do label encoding for team1, team2, toss_winner, toss_decision, winner and mention the encoding in the comments
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
toss_mapping = {}

for column in df1.columns:
    df1[column] = le.fit_transform(df1[column])
    toss_mapping[column] = dict(zip(le.classes_, le.transform(le.classes_)))

toss_mapping
```

```python
df1.head()
```

```python
X=df1.drop('toss_decision',axis='columns')
y=df1['toss_decision']
```

```python
X.head()
```

```python
#split the data into train and test set

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from tensorflow import keras
import tensorflow as tf
keras = tf.keras
layers = keras.layers
EarlyStopping = keras.callbacks.EarlyStopping


model = keras.Sequential([
    layers.Input(shape=(X_train.shape[1],)),  # Input layer
    layers.Dense(8,activation='selu'),
    layers.Dense(1, activation='sigmoid')    # Output layer with 1 unit and sigmoid activation

])

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics='accuracy')

# Define early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10)

# Train the model with early stopping
history = model.fit(X_train, y_train, epochs=200, batch_size=4, validation_data=(X_test, y_test), callbacks=[early_stopping])
```

```python
#print the accuracy score for the model
print(model.evaluate(X_test, y_test))
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import GridSearchCV

log_reg = LogisticRegression()

param_grid = {'C': [0.001, 0.01, 0.1, 1, 10, 100]}
log_reg_cv = GridSearchCV(log_reg, param_grid, cv=5)
log_reg_cv.fit(X_train, y_train)

print("Tuned Logistic Regression Parameters: {}".format(log_reg_cv.best_params_))
print("Best score is {}".format(log_reg_cv.best_score_))
```

```python
from sklearn.svm import SVC
model = SVC(kernel='rbf', C=1, verbose=1)
model.fit(X_train, y_train)
print(model.score(X_test, y_test))
```
