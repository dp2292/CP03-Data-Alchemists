```python
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('stadium_details.csv')
```

```python
plt.figure(figsize=(4, 4))
labels = ['Batting First', 'Chasing']
sizes = [df['won_after_bat_first'].sum(), df['won_after_chase'].sum()]
colors = ['lightcoral', 'lightskyblue']
plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%', startangle=140)
plt.title('Distribution of Matches Won')
plt.show()
```

```python
# Bar plot for 'venue'
plt.figure(figsize=(6,6))
sns.countplot(y='venue', data=df)
plt.title('Number of Matches per Venue')
plt.xlabel('Count')
plt.ylabel('Venue')
plt.show()

# Pie chart for 'won_after_bat_first'
plt.figure(figsize=(6,6))
df['won_after_bat_first'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Win Percentage After Batting First')
plt.show()

# Pie chart for 'won_after_chase'
plt.figure(figsize=(6,6))
df['won_after_chase'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Win Percentage After Chasing')
plt.show()

# Histogram for 'first_inning_score'
plt.figure(figsize=(6,6))
sns.histplot(df['first_inning_score'], bins=30, kde=True)
plt.title('Distribution of First Inning Scores')
plt.xlabel('First Inning Score')
plt.ylabel('Frequency')
plt.show()

# Histogram for 'second_inning_score'
plt.figure(figsize=(6,6))
sns.histplot(df['second_inning_score'], bins=30, kde=True)
plt.title('Distribution of Second Inning Scores')
plt.xlabel('Second Inning Score')
plt.ylabel('Frequency')
plt.show()

# Boxplot for 'first_inning_score'
plt.figure(figsize=(6,6))
sns.boxplot(x='first_inning_score', data=df)
plt.title('Boxplot of First Inning Scores')
plt.xlabel('First Inning Score')
plt.show()

# Boxplot for 'second_inning_score'
plt.figure(figsize=(6,6))
sns.boxplot(x='second_inning_score', data=df)
plt.title('Boxplot of Second Inning Scores')
plt.xlabel('Second Inning Score')
plt.show()
```

```python
venues = df['venue'].unique()

# For each venue
for venue in venues:
    # Filter DataFrame for the current venue
    df_venue = df[df['venue'] == venue]
    # Calculate total matches
    total_matches = df_venue['won_after_bat_first'] + df_venue['won_after_chase']

    # Calculate win percentage after batting first
    win_percentage_bat_first = df_venue['won_after_bat_first'] / total_matches * 100

    # Pie chart for 'won_after_bat_first'
    plt.figure(figsize=(6,6))
    plt.pie([win_percentage_bat_first, 100-win_percentage_bat_first], labels=['Won After Batting First', 'Lost or Draw'], autopct='%1.1f%%')
    plt.title(f'Win Percentage After Batting First at {venue}')
    plt.show()
```
